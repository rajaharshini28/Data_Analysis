### Empirical Rule

It is a quick way to estimate the "SPREAD OF DATA IN A NORMAL DISTRUBUTION" BASED ON The standarad deviation

->68% within 1 standard deviation of mean

->95% within 2 standard deviation of mean

->99.7% within 3 standard deviation of mean

![alt text](image.png)


### Distortion in Normal distrubution

> Skewness:-

    -> for left skew-mean and median are less than mode 
    -> for right skew-mean and median are more than mode
    -> for nomral distrubution-mean,median and mode are  lying in same place at the center

>Kurtosis:-
    
    -> Mesokurtic: Tails are of similar weigth to a normal distrubution
            > Kurtosis ~=3

    ->Lepokurtic: Heavier tails than a normal distrubution 
            > Kurtosis:>3
